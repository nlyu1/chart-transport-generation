```python
from __future__ import annotations

import torch
from IPython.display import display

from src.model.mlp import StackedResidualMLPConfig
from src.model.time_conditioning import (
    CategoricalConditioningConfig,
    TimeConditioningConfig,
)

torch.manual_seed(0)
```

### Plain StackedResidualMLPConfig

```python
plain_config = StackedResidualMLPConfig.initialize(
    layer_dims=[4, 8, 6],
)

display(plain_config.visualize())
```

```python
plain_model = plain_config.get_model()
plain_x = torch.randn(3, plain_config.layer_dims[0])
plain_output = plain_model(plain_x)

print(f"plain: {tuple(plain_x.shape)} -> {tuple(plain_output.shape)}")
```

### Time And Categorical Conditioning

```python
conditioned_config = StackedResidualMLPConfig.initialize(
    layer_dims=[4, 8, 6],
    time_conditioning_config=TimeConditioningConfig(
        min_t_lambda=1e-3,
        max_t_lambda=1.0,
        condition_dim=8,
    ),
    cat_conditioning_config=CategoricalConditioningConfig(
        num_classes=3,
        condition_dim=4,
    ),
)

display(conditioned_config.visualize())
```

```python
conditioned_model = conditioned_config.get_model()
conditioned_x = torch.randn(3, conditioned_config.layer_dims[0])
t = torch.linspace(0.0, 1.0, conditioned_x.shape[0])
categorical = torch.tensor([0, 1, 2])
conditioned_output = conditioned_model(
    conditioned_x,
    t=t,
    categorical=categorical,
)

print(
    f"conditioned: {tuple(conditioned_x.shape)} -> {tuple(conditioned_output.shape)}"
)
```
