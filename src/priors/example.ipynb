```python
import torch
from IPython.display import display

from src.priors.anchored import AnchoredGaussianScaleMixturePriorConfig
from src.priors.unimodal import UnimodalGaussianPriorConfig

unimodal_prior = UnimodalGaussianPriorConfig.initialize(latent_shape=[2])
anchored_prior = AnchoredGaussianScaleMixturePriorConfig.initialize(
    latent_shape=[2],
    precision=4.0,
)

display(unimodal_prior.visualize())
display(anchored_prior.visualize())
```

```python
from src.priors.utils import plot_prior_with_scores

sample_count = 3000
subset_size = 50

unimodal_samples = unimodal_prior.sample(batch_size=sample_count)
unimodal_indices = torch.randperm(unimodal_samples.shape[0])[:subset_size]
unimodal_subset = unimodal_samples[unimodal_indices]
unimodal_scores = unimodal_prior.analytic_score(
    y_t=unimodal_subset,
    t=torch.zeros(subset_size),
)

fig = plot_prior_with_scores(
    samples=unimodal_samples,
    anchor_samples=unimodal_subset,
    scores=unimodal_scores,
    title="Unimodal Gaussian Prior With Analytic Score Vectors at t=0",
    sample_trace_name="samples",
    arrow_trace_name="analytic_score at t=0",
    sample_color="rgb(31, 119, 180)",
    alpha=0.2,
    arrow_color="crimson",
    arrow_scale=0.2,
)
fig.show()
```

```python
anchored_samples = anchored_prior.sample(batch_size=sample_count)
anchored_indices = torch.randperm(anchored_samples.shape[0])[:subset_size]
anchored_subset = anchored_samples[anchored_indices]
anchored_scores = anchored_prior.analytic_score(
    y_t=anchored_subset,
    t=torch.zeros(subset_size),
)

fig = plot_prior_with_scores(
    samples=anchored_samples,
    anchor_samples=anchored_subset,
    scores=anchored_scores,
    title="Anchored Scale-Mixture Prior With Analytic Score Vectors at t=0",
    sample_trace_name="samples",
    arrow_trace_name="analytic_score at t=0",
    sample_color="rgb(44, 160, 44)",
    alpha=0.2,
    arrow_color="darkorange",
    arrow_scale=0.2,
)
fig.show()
```
